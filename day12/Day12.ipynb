```python
%matplotlib inline
import networkx as nx
```

```python
# Read edges and initialize the graph
f = open("input1.txt", 'r')

data = f.read()

Lin = data.split('\n')

# Initialize the graph
G = nx.Graph()

# Add the edges
for l in Lin:
    if l != '':
        a, b = l.split('-')
        G.add_edge(a, b)

nx.draw_circular(G, with_labels=True)
```

### First Part

Each path will be a list with:
* 0: Current node
* 1: Set of the small caves visited

```python
def solve1():
    open_path = list()
    open_path.append(['start', list()])

    paths = 0

    while len(open_path) != 0:
        curr_node, small_caves = open_path.pop()

        if curr_node == 'end':
            paths += 1

        else:
            for neighbor_node in list(G.adj[curr_node]):
                if neighbor_node not in small_caves:
                    if curr_node.islower():
                        open_path.append((neighbor_node, small_caves+[curr_node]))
                    else:
                        open_path.append((neighbor_node, small_caves))
    
    return paths
```

```python
solve1()
```

### Second Part

Each path will be a list with:
* 0: Current node
* 1: Set of the small caves visited
* 2: Bool indicating if we **have** the *power* of visiting a small cave twice

```python
def solve2():
    open_path = list()
    open_path.append(['start', ['start'], True])

    paths = 0

    while len(open_path) != 0:
        curr_node, small_caves, power = open_path.pop()
        
        if curr_node == 'end':
            paths += 1

        else:
            for neighbor_node in list(G.adj[curr_node]):
                if neighbor_node not in small_caves:
                    if neighbor_node.islower():
                        open_path.append((neighbor_node, small_caves + [neighbor_node], power))
                    else:
                        open_path.append((neighbor_node, small_caves, power))
                else:
                    if power and neighbor_node != 'start':
                        open_path.append((neighbor_node, small_caves, False))
        
    return paths
```

```python
solve2()
```
