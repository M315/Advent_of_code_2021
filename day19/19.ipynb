The rotations correspond to the porduct of S3, the permitations of 3 elements, and the vectors of 3 dimentions of Z mod 2.  
So for all the directions of the axis we have all the possible permutations of them.

```python
def gen_orientations():
    # 48
    for i in (-1, 1):
        for j in (-1, 1):
            for k in (-1, 1):
                for rest in itertools.permutations(range(3)):
                    yield rest + (i, j, k)

def rotation(orientation, point):
    x, y, z, dx, dy, dz = orientation
    ret = [None] * 3
    ret[x] = point[0] * dx
    ret[y] = point[1] * dy
    ret[z] = point[2] * dz
    
    return tuple(ret)

def add(a, b):
    ret = []
    
    for i in range(len(a)):
        ret.append(a[i] + b[i])
        
    return tuple(ret)

def sub(a, b):
    ret = []
    
    for i in range(len(a)):
        ret.append(a[i] - b[i])
        
    return tuple(ret) 
```

```python
def solve(scanners):
    # Set the locations and orientations of the scanners to NULL
    location = [None] * len(scanners)
    orientation = [None] * len(scanners)
    
    # Set the location and the orientation of the scanner 0 to origin and identity
    location[0] = (0, 0, 0)
    orientation[0] = (0, 1, 2, 1, 1, 1)
    
    beacons = set()
    
    # Set of the pairs of scanners that are not correlated
    bad = set()
    
    while True:
        for i in range(len(scanners)):
            # Use the known scanners
            if location[i] is None:
                continue
            
            # Set the points to the relative possition from scanner 0
            i_real = []
            for p in scanners[i]:
                i_real.append(add(location[i], rotation(orientation[i], p)))
            i_real_set = set(i_real)
        
            for j in range(len(scanners)):
                # Check that the candidates are not the same
                # That the second is unknow and are not uncorrelated
                if i == j or location[j] is not None or (i, j) in bad:
                    continue
                
                #print('trying', i, j)
                for ori in gen_orientations():
                    # Compute the new relative positions after the rotation
                    rel_pos = []
                    for p in scanners[j]:
                        rel_pos.append(rotation(ori, p))
                    
                    # Count the intersections for each point of the known scanner
                    for ki in range(len(i_real)):
                        for kj in range(len(rel_pos)):
                            # Assume that kj is at the same real possition as ki
                            traslation = sub(i_real[ki], rel_pos[kj])
                            
                            count = 0
                            for k in range(len(rel_pos)):
                                pos = add(rel_pos[k], traslation)
                                if pos in i_real_set:
                                    count += 1
                                    if count >= 12:
                                        break
                            
                            # Found!!
                            if count >= 12:
                                print('found', j)
                                location[j] = traslation
                                orientation[j] = ori
                                break
                        
                        if location[j] is not None:
                            break

                    if location[j] is not None:
                        break

                if location[j] is None:
                    bad.add((i, j))

                else:
                    break
            
        if all(location):
            break
    
    for i in range(len(scanners)):
        for p in scanners[i]:
            beacons.add(add(location[i], rotation(orientation[i], p)))
    
    M = 0
    for p1 in location:
        for p2 in location:
            d = sub(p1, p2)
            M = max(abs(d[0]) + abs(d[1]) + abs(d[2]), M)
    
    return M
```
